# Making it a function

```python
from nltk.tokenize import word_tokenize
def word2vec(x):
    # Making everything lowercase
    x = str(x)
    x = x.lower()

    # Tokenizing it
    tokenized_list = word_tokenize(x)
    
    temp = []
    for i in tokenized_list:
        try:
            temp.append(model[i])
        except:
            pass
    
    if len(temp) == 0:
        result = 0
    else:    
        result = sum(temp)/len(temp)
    
    return result
```

```python
# Let's try the funciton
word2vec('')
```
